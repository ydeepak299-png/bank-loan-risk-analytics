```python
# ==========================================================
# Unique Values in Categorical Columns
# ==========================================================

categorical_columns = [
    "gender",
    "education",
    "home_ownership",
    "loan_purpose",
    "previous_defaults",
    "loan_status"
]

for column in categorical_columns:
    print(f"\n{column}")
    print(df[column].value_counts(dropna=False))
    print("-" * 50)
```

```python
# ==========================================================
# Numerical Range Validation
# ==========================================================

numerical_columns = [
    "age",
    "income",
    "employment_experience",
    "loan_amount",
    "interest_rate",
    "loan_to_income_ratio",
    "credit_history_years",
    "credit_score"
]

range_summary = df[numerical_columns].agg(["min", "max", "mean", "median"]).T

range_summary
```

```python
# ==========================================================
# Loan Status Distribution
# ==========================================================

loan_status_summary = df["loan_status"].value_counts().sort_index()

loan_status_percentage = (
    df["loan_status"]
    .value_counts(normalize=True)
    .sort_index()
    .mul(100)
    .round(2)
)

print("Loan Status Count:")
print(loan_status_summary)

print("\nLoan Status Percentage:")
print(loan_status_percentage)
```

```python
# ==========================================================
# Create Readable Loan Status Label
# ==========================================================

df["loan_status_label"] = df["loan_status"].map({
    0: "Non-Default",
    1: "Default"
})

df[["loan_status", "loan_status_label"]].head()
```

```python
# ==========================================================
# Check Text Columns for Extra Spaces
# ==========================================================

text_columns = df.select_dtypes(include="object").columns

print("Text Columns:")
print(text_columns)
```

```python
# ==========================================================
# Remove Leading and Trailing Spaces
# ==========================================================

for col in text_columns:
    df[col] = df[col].str.strip()

print("Extra spaces removed successfully.")
```

```python
# ==========================================================
# Standardize Text Columns
# ==========================================================

for col in text_columns:
    df[col] = df[col].str.title()

print("Categorical values standardized.")
```

```python
# ==========================================================
# Verify Standardized Values
# ==========================================================

for col in text_columns:
    print(f"\n{col}")
    print(df[col].unique()[:10])
```

```python
# ==========================================================
# Improve Category Names
# ==========================================================

df["loan_purpose"] = df["loan_purpose"].replace({
    "Homeimprovement": "Home Improvement",
    "Debtconsolidation": "Debt Consolidation"
})

print(df["loan_purpose"].unique())
```

```python
# ==========================================================
# Section 6 : Outlier Detection
# ==========================================================

import matplotlib.pyplot as plt

numeric_columns = [
    "age",
    "income",
    "employment_experience",
    "loan_amount",
    "interest_rate",
    "loan_to_income_ratio",
    "credit_history_years",
    "credit_score"
]

for col in numeric_columns:
    plt.figure(figsize=(8,2))
    plt.boxplot(df[col], vert=False)
    plt.title(f"Boxplot - {col}")
    plt.show()
```

```python
# ==========================================================
# IQR Outlier Summary
# ==========================================================

outlier_summary = []

for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    outlier_summary.append({
        "Column": col,
        "Outliers": len(outliers)
    })

outlier_df = pd.DataFrame(outlier_summary)

outlier_df
```

## Outlier Decision

Outliers were identified using the IQR method.

After reviewing the business context, the outliers were retained because they represent valid customer behavior rather than data entry errors. High-income customers, large loan amounts, and strong credit scores are expected in a real banking portfolio.

Therefore, no outliers were removed from the dataset.
